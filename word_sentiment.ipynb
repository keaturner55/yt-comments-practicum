```python
from nltk.sentiment import SentimentIntensityAnalyzer
import operator
import sqlite3
import os
import pandas as pd
```

```python
DATA_DIR = 'C:\\Users\\keatu\\Regis_archive\\practicum2_data\\'
dbname = os.path.join(DATA_DIR, "Youtube_Data_proof_concept.db")
```

```python
def get_sql_table(table, db_name):
    con = sqlite3.connect(db_name)
    df = pd.read_sql("select * from {}".format(table),con)
    return df
```

```python
commentsdf = get_sql_table('comments', dbname)
```

```python
sia = SentimentIntensityAnalyzer()
```

```python
commentsdf['sentiment_score'] = commentsdf["textOriginal"].apply(lambda x: sia.polarity_scores(x)['compound'])
```
